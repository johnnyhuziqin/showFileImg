```python
def  showFileImage(fileList):
    num_pic = len(fileList)
    x_id = 0
    y_id = 0
    num_x = 4
    num_y = (num_pic+num_x-1)/num_x

    fig, axs = plt.subplots(num_y, num_x,figsize=(20,num_pic*2))
    plt.figure(figsize=(20,60))
    for pic_id in range(num_pic):
        if x_id >= num_x:
            x_id = 0
            y_id = y_id + 1
        axs_n = axs[y_id,x_id] 
        path = fileList[pic_id]
        image = plt.imread(path)
        axs_n.imshow(image)
        #axs_n.axis('off')
        fileName = os.path.split(path)[1]
        axs_n.set_title(fileName,fontsize=10)
        imgSize = getImgSize(path)
        axs_n.set_xlabel('size:{}'.format(imgSize))
        x_id = x_id + 1

def showDirPics(path):
    fileListInDir = GetFileList(path,[])
    print("total files:%d"%(len(fileListInDir)))
    showFileImage(fileListInDir) 
    
#picPath1 = "/home/huziqin/work/gitwork/8388/application/customer_app/siba/target/siba/PA/Resource/home/"    
#showDirPics(picPath2)
```
